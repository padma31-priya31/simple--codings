# import all libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
from sklearn.cluster import KMeans
```

# read data 

```python
data =pd.read_csv(r'D:\Program Files\iris.csv')
data.head()
```

# Visualization of data 

```python
data.plot()
```

```python
data.boxplot()
```

# Preparing data

```python
X = data.iloc[:,[0,1,2,3]].values
X
```

# Now plotting elbow method to identify numbers of cluster

```python
K = range(1,11)
Wcss =[]
for i in K:
    model =KMeans(n_clusters = i,init = 'k-means++',n_init = 10,max_iter = 300,random_state =0)
    model.fit(X)
    Wcss.append(model.inertia_)
plt.plot(K,Wcss)
plt.title('elbow method')
plt.xlabel('cluster range')
plt.ylabel('Wcss')
plt.show()

    
```

```python
model =KMeans(n_clusters = 3,init = 'k-means++',n_init = 10,max_iter = 300,random_state =0)
y_kmeans =model.fit_predict(X)
print(y_kmeans)
```

# Classication report and Accuracy score for data

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
predictions = model.predict(X)
print(predictions)
print(classification_report(y_kmeans,predictions))
print(accuracy_score(y_kmeans,predictions))
```

# visualizing the clusters 

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s =100,c ='black',label='iris.setosa')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s =100,c ='blue',label='iris.versicolour')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s= 100,c ='red',label='iris.virginica')

plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],
           s =100,c ='violet',label ='centroids')
plt.legend()
plt.show()
```
